### DESAFIO

Sucesso! A pizzaria do bairro tornou-se uma multinacional que engoliu todas as concorrentes. Querendo crescer ainda mais, o dono sugere um novo recurso: criar sua própria pizza. Para isso, temos um dicionário com todos os ingredientes e seus respectivos preços. A pizzaria recebeu os seguintes pedidos:
-  1: Massa sem glúten, molho barbecue, frango, pimentão e tomate
-  2: Massa tradicional, molho de tomate, atum, mussarela, azeitona, cebola, pimentão, tomate e borda com queijo
-  3: Massa integral, molho bechamel, cebola, pimentão, tomate, frango e borda com requeijão

Imprima uma lista com todos os itens do pedido e seu preço final que será retornado ao cliente. Armazene o valor das vendas de forma que não seja possível alterar seus valores e imprima.

```python
from collections import namedtuple

# Inicializa a namedtuple 'vendas'
vendas = namedtuple('vendas', ['total'])
vendas = vendas(total=0.0)

# Dicionário com os preços dos ingredientes
ingredientes_preços = {
    "massa sem gluten": 10.0,
    "massa tradicional": 9.0,
    "massa integral": 11.0,
    "molho barbecue": 2.0,
    "molho de tomate": 1.5,
    "molho bechamel": 3.0,
    "frango": 8.0,
    "atum": 5.0,
    "mussarela": 3.0,
    "azeitona": 1.5,
    "cebola": 1.0,
    "pimentão": 1.5,
    "tomate": 1.0,
    "borda com queijo": 2.0,
    "borda com requeijão": 2.5,
    "refrigerante": 5.0  # Adicione o preço do refrigerante
}

def cal_preç_ped(pedido):
    preço_total = 0.0
    for item in pedido:
        if item in ingredientes_preços:
            preço_total += ingredientes_preços[item]
        else:
            print(f'item desconhecido {item}')
    return preço_total

# Cria um dicionário vazio para os pedidos de pizza
pedidos_pizza = {}

while True:
    chave = input('Deseja pizza? Digite nome: ')
    ingredientes = input('Digite os ingredientes separados por vírgula: ').lower()
    criando_lista = ingredientes.split(',')
    
    pedidos_pizza[chave] = criando_lista
    
    continuar = input('Deseja fazer outro pedido de pizza? (sim ou não): ')
    if continuar.lower() != 'sim':
        break

# Cria um dicionário vazio para os pedidos de refrigerante
pedidos_refrigerante = {}

while True:
    quantidade_refrigerante = int(input('Quantos refrigerantes deseja adicionar? (0 se nenhum): '))
    if quantidade_refrigerante > 0:
        pedidos_refrigerante['refrigerante'] = ['refrigerante'] * quantidade_refrigerante
    
    # continuar = input('Deseja fazer outro pedido de refrigerante? (sim ou não): ')
    # if continuar.lower() != 'sim':
        break

# Calcula o preço dos pedidos de pizza
preço_total_pizza = 0.0
for pedido in pedidos_pizza.values():
    preço_pedido_pizza = cal_preç_ped(pedido)
    preço_total_pizza += preço_pedido_pizza

# Calcula o preço dos pedidos de refrigerante
preço_total_refrigerante = 0.0
for pedido in pedidos_refrigerante.values():
    preço_pedido_refrigerante = cal_preç_ped(pedido)
    preço_total_refrigerante += preço_pedido_refrigerante

# Imprime os totais separadamente
print(f'pedido : {chave}')
print(f'Total do pedido de pizza :{criando_lista} R$ {preço_total_pizza:.2f}')
print(f'Total {quantidade_refrigerante}  refrigerante: R$ {preço_total_refrigerante:.2f}')

# Atualiza o total de vendas
vendas = vendas._replace(total=vendas.total + preço_total_pizza + preço_total_refrigerante)
print(f'Total da venda: R$ {vendas.total:.2f}')
```
